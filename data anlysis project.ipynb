# 5.Data Visualization

#### 1. what was the best year of sales (How many sales)

```python
year_group = df.groupby('year')['item_id'].count()
year_group
```

```python
plt.bar(year_group.index, year_group.values, color='indigo')
plt.plot(year_group.index, year_group.values)
plt.xlabel('Years')
plt.ylabel('Number of sales')
plt.title('Number of sales happened in each year')
```

```python
plt.plot(year_group.index, year_group.values)
```

#### 2. what was the best month of sales (How many sales)

```python
month_group = df.groupby('month')['item_id'].count()
month_group
```

```python
plt.plot(month_group.index, month_group.values, color='indigo')
plt.xlabel('Months')
plt.ylabel('Number of sales')
plt.title('Number of sales happened in each month')
```

#### 3. Group by year and calculate total sales

```python
year_group = df.groupby('year')['amount'].sum()
year_group
```

```python
plt.bar(year_group.index, year_group.values, color='indigo')
plt.plot(year_group.index, year_group.values, color='red')
plt.xlabel('Years')
plt.ylabel('Total sales')
plt.title('Total sales happened in each year')
```

#### 4. Top 10 brands sold the most in 2015 to 2018

```python
filter = (df['year']>=2015) & (df['year']<=2018)
df_15_to_18 = df[filter]
df_15_to_18
```

```python
brand_group = df_15_to_18.groupby('brand')['amount'].sum().nlargest(10)
brand_group
```

```python
plt.bar(brand_group.index, brand_group.values, color='indigo')
plt.xlabel('Brand')
plt.ylabel('Total sales')
plt.title('Top 10 Brands based on total sales')
```

#### 5. Create different plots for each year 2016,2017,2018 for top 5 brands sold the most

```python
# filtering data for 2016, 2017 and 2018
df_2016 = df[df['year']==2016]
df_2017 = df[df['year']==2017]
df_2018 = df[df['year']==2018]

# creating brand groups for 2016, 2017 and 2018
brand_group_2016 = df_2016.groupby('brand')['amount'].sum().nlargest(5)
brand_group_2017 = df_2017.groupby('brand')['amount'].sum().nlargest(5)
brand_group_2018 = df_2018.groupby('brand')['amount'].sum().nlargest(5)

# creating subplots
fig, axs = plt.subplots(1,3, figsize=(15,4))

# plot for first subplot
axs[0].bar(brand_group_2016.index, brand_group_2016.values, color='indigo')
axs[0].set_title('Top 5 brands in 2016')

# plot for first subplot
axs[1].bar(brand_group_2017.index, brand_group_2017.values, color='green')
axs[1].set_title('Top 5 brands in 2017')

# plot for first subplot
axs[2].bar(brand_group_2018.index, brand_group_2018.values)
axs[2].set_title('Top 5 brands in 2018')
```

#### 6. What category sold the most in 2015 to 2018?

```python
filter = (df['year']>=2015) & (df['year']<=2018)
df_2015_to_2018 = df[filter]
category_group = df_2015_to_2018.groupby('category')['amount'].sum().nlargest(10)

plt.bar(category_group.index, category_group.values, color='indigo')
plt.xlabel('Category')
plt.ylabel('Total sales')
plt.xticks(rotation=90)
plt.title('Top 10 Categories based on total sales')
```

#### 7. The distribution of ratings

```python
x = df['rating'].value_counts()

plt.bar(x.index, x.values, color='indigo')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Frequency of different ratings')
```

#### 8. What is the most rated brand name  between 2015 to 2018?

```python
filter = (df['year']>=2015) & (df['year']<=2018)
df_2015_to_2018 = df[filter]
brand_group = df_2015_to_2018.groupby('brand')['rating'].mean().nlargest(10)

plt.bar(brand_group.index, brand_group.values, color='indigo')
plt.xlabel('Brand')
plt.ylabel('Average rating')
plt.xticks(rotation=90)
plt.title('Top 10 Brand based on ratings')
```

#### 9. category percentage sales

```python
category_group = df.groupby('category')['amount'].sum().nlargest(5)

plt.pie(category_group.values, labels=category_group.index, autopct="%.2f%%")
plt.title('Top 10 Categories with highest percentage of sales')
plt.show()
```

#### 11. brand wise sales percentage


```python
brand_group = df.groupby('brand')['amount'].sum().nlargest(5)

plt.pie(brand_group.values, labels=brand_group.index, autopct="%.2f%%")
plt.title('Top 10 Brands with highest percentage of sales')
plt.show()
```

#### 12. Gender wise customer distribution

```python
gender_distribution = df['gender'].value_counts()
plt.pie(gender_distribution.values, labels=gender_distribution.index, autopct='%1.1f%%')
plt.title('Gender wise customer Distribution')
plt.show()
```

#### 13. Group by gender and category, then calculate the total quantity sold in different charts

```python
gender_category = df.groupby(['gender', 'category'])['quantity'].sum()

# Plot a pie chart for each gender
genders = df['gender'].unique()

for gender in genders:
    plt.pie(gender_category[gender], labels=gender_category[gender].index, autopct='%.1f%%')
    plt.title(f'{gender} - Product Preferences')
    plt.show()
```
